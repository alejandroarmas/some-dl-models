```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset
train = pd.read_csv('stage_2_data/train.csv')
test = pd.read_csv('stage_2_data/test.csv')
```

## Dataset description

```python
train.describe()
```

```python
test.describe()
```

```python
trn = np.unique(train['5'], return_counts=True)
```

```python
tst = np.unique(test['7'], return_counts=True)
```

```python
plt.bar(trn[0], trn[1])
plt.xticks(trn[0])
plt.title("Train Label Distribution")
plt.xlabel("Digit/Class")
plt.ylabel("Frequency")
```

```python
plt.bar(tst[0], tst[1])
plt.xticks(trn[0])
plt.title("Test Label Distribution")
plt.xlabel("Digit/Class")
plt.ylabel("Frequency")
```

## Dataset preprocessing

```python
# Rename label columns
train.rename(columns={'5': 'label'}, inplace=True)
test.rename(columns={'7': 'label'}, inplace=True)
```

```python
# Save new datasets
train.drop(['Unnamed: 0'], axis=1).to_csv('stage_2_data/train.csv', index=False)
test.to_csv('stage_2_data/test.csv', index=False)
```
