```python
# Imports
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

## MNIST

```python
# MNIST
f = open('stage_3_data/stage_3_data/MNIST', 'rb')
data = pickle.load(f)
f.close()
print('training set size:', len(data['train']), 'testing set size:', len(data['test']))
train_dict = np.array(data['train'])
test_dict = np.array(data['test'])
X_train = []
y_train = []
X_test = []
y_test = []
for instance in train_dict:
    img = instance['image']
    label = instance['label']
    X_train.append(img)
    y_train.append(label)
for instance in test_dict:
    img = instance['image']
    label = instance['label']
    X_test.append(img)
    y_test.append(label)
```

```python
trn = np.unique(y_train, return_counts=True)
plt.bar(trn[0], trn[1])
plt.xticks(trn[0])
plt.title("Train Label Distribution")
plt.xlabel("Digit/Class")
plt.ylabel("Frequency")
```

```python
tst = np.unique(y_test, return_counts=True)
plt.bar(tst[0], tst[1])
plt.xticks(trn[0])
plt.title("Test Label Distribution")
plt.xlabel("Digit/Class")
plt.ylabel("Frequency")
```
